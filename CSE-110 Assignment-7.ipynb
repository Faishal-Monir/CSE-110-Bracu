```python
#TASK-1                                #input : #Bubble sort 
my_list=[10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
x=my_list.copy()
store=x
for i in range (len(x)-1):
    for j in range (0,(len(x)-i-1)):
        if x[j]>x[j+1]:
            temp=x[j]
            x[j]=x[j+1]
            x[j+1]=temp
 
print(store)
```

```python
#TASK-2                                #input : #selection Sort
my_list= [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
data=my_list.copy()
for i in range (len(data)):
    min=data[i]
    min_idx=i
    for j in range(i+1,len(data)):
        if data[j]<min:
            min=data[j]
            min_idx=j
        
    temp=data[min_idx]
    data[min_idx]=data[i]
    data[i]=temp
 
print(data) 
```

```python
#TASK-5                                #input :MAT110
def mns(marks,course):
    idx=0
    if course=="CSE110":
        idx=1
    elif course=="PHY111":
        idx=2
    elif course=="MAT110":
        idx=3
    else:
        return "Wrong Input"
    
#Used bubble sort!
    for i in range (len(marks)-1):
        for j in range (0,(len(marks)-i-1)):
            if marks[j][idx] < marks[j+1][idx]:
                temp=marks[j]
                marks[j]=marks[j+1]
                marks[j+1]=temp
    print()
    print("----OUTPUT----")
    for k in marks:
        print(k[0])
lst_a = [["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90]] 
course_a= input("Enter your course name:")
mns(lst_a,course_a)
```

```python
#TASK-8                                #input :
list_long=int(input("Enter list length:"))
list_one=[]
for i in range(list_long):
    number=int(input("Input list elements:"))
    list_one.append(number)
def calc (lst,size):
    min_a=0
    min_b=1
    summation=lst[0]+lst[1]
    for j in range(0,size-1):
        for k in range(j+1,size):
            sum_b=lst[j]+lst[k]
            if abs(summation)>abs(sum_b):
                summation=sum_b
                min_a=j
                min_b=k
    print("----OUTPUT----")
    print()
    print(f"Two pairs which have the smallest sum = {lst[min_a]} and {lst[min_b]}")
 
calc(list_one,list_long)
```

```python
#TASK-9                                #input :
import math
points =[(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]
x=points
copylst=x.copy()
store={}
for i in copylst:
    point1,point2=i
    calc=math.sqrt((point1**2)+(point2**2))
    store.update({i:calc})
 
final_a=sorted(store.values())
final={}
for j in final_a:
    for k in store.keys():
        if store[k]==j:
            final[k]=store[k]
keys=[]
values=[]
for l in final.keys():
    keys.append(l)
 
for m in final.values():
    values.append(m)
print("Minimum distance =",values[0])
print("Here the closest point is",keys[0],"which has a distance of",values[0],"from the origin.")
```
