```python
#TASK-6                               #input : "7, 13, 2, 10, 6, -11, 0"
data=input("Please enter string seperated by comas(,):")
data_a=data[1:-1]
dataset=[]
dataset=data_a.split(",")
for j in range (len(dataset)):
    dataset[j]=int(dataset[j])
 
big=dataset[0]
for k in dataset:
    if k>=big:
        big=k 
 
y=dataset.copy()
position=y.index(big) 
print("________________ output ___________________")
print()
print("My list:",dataset) 
print("Largest number in the list is",big,"which was found at index",position)
```

```python
#TASK-9                                #input : 7 12 4 55 96 2 11 61 33 42
data=input("Enter your numbers seperated by spaces:")
x="0123456789"
y=[]
temp=""
for i in data:
    if i in x:
        temp=temp+i
    else:
        y.append(temp)
        temp=""
y.append(temp)
a=[]
for k in y:
    if k==",":
        pass
    else:
        a.append(int(k))
 
z=[]
for j in y:
    if int(j)%2!=0:
        z.append(int(j))
    else:
        pass
 
print("________________ output ___________________")
print()
print("Original list:",a) 
print("Modified list:",z)
```
