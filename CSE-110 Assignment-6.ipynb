```python
#TASK-7                                #input :
def show_palindrome(x):
    first_part=""
    second_part=""
    
    for i in range (1,x,1):
        first_part+=str(i)
    for j in range (x,0,-1):
        second_part+=str(j)
    
    final=first_part+second_part
    return final
 
print(show_palindrome(5))
```

```python
#TASK-8                                #input :
def show_palindrome(x):
    first_part=""
    second_part=""
    
    for i in range (1,x,1):
        first_part+=" "+str(i)
    for j in range (x,0,-1):
        second_part+=" "+str(j)
    
    final=first_part+second_part
    return final
def show_palindromic_triangle(x):
    gap=(x-1)*2
    for i in range(1,x+1):
        print(" "*gap,end="")
        gap-=2
        print(show_palindrome(i))
    
show_palindromic_triangle(5)
```

```python
#TASK-11                                #input :
def rem_duplicate(data):
    final=[]
    for i in data:
        if i not in final:
            final.append(i)
        else:
            pass
    final=tuple(final)
    return final
 
print(rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0)))
```

```python
#TASK-12                                #input :
def dup_sus(data):
    final=[]
    remove=0
    for i in data:
        if final.count(i)<2:
            final.append(i)
        else:
            remove+=1
    print("Removed:",remove)
    return final
dup_sus([1, 2, 3, 3, 3, 3, 4, 5, 8, 8]) 
```

```python
#TASK-13                                #input :
def calc(operation,data,data_a):
    if operation=="+":
        return (data + data_a)
    elif operation=="-":
        return(data - data_a)
    elif operation=="*":
        return(data * data_a) 
    elif operation=="/":
        return(data / data_a)
operations=input("Enter operation:")
datas=float(input("Enter first number:"))
datas_a=float(input("Enter second number:"))
print(calc(operations,datas,datas_a))
```

```python
#TASK-14                                #input : I love programming.,3
def index_remover(sentence,position):
    newline=sentence[0]
    lastpart=""
    for i in range (1,len(sentence)):
        if i%position !=0:
            newline+=sentence[i]
        else:
            lastpart+=sentence[i]
    newline+=lastpart
    return newline
line=input("Enter your line:")
pos=int(input("Enter position:"))
print(index_remover(line,pos))
```
