```python
#TASK-8                                #input : {'Jon': 100, 'Dan':200, 'Rob':30, 'Ned':110}

data=input("Enter your dictionary:")
data=data[1:-1]
convert=data.split(",")
final={}
for i in convert:
    final[i.split(":")[0]]=int(i.split(":")[1])
 
summation=0
counter=0 
for j in final.values():
    summation+=j
    counter+=1
average=int(summation/counter)
print("Average is",average)
```
